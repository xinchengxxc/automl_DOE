```python
import os
import csv
experiments=[]
experiments.append("Model_dir,Traindata_name,Testdata_name")
# 指定路径
folder_path = r'C:\Users\xuxuk\Desktop\xxk\DOE_paper\code_organized\DOE_samping\Modelbase'
# folder_path= folder_path +mymodel
# 遍历文件夹中的所有文件
for filename in os.listdir(folder_path):

    model_path= folder_path +"\\"+ filename
    for datasetname in os.listdir(model_path):
        # 检查文件是否以 '.csv' 结尾
        if datasetname.endswith('.csv') and datasetname !="LHD_data_100000.csv":
            # 如果是，打印文件名
            print(filename+","+datasetname+","+"LHD_data_200000.csv")
            experiments.append(filename+","+datasetname+","+"LHD_data_200000.csv")
            
# 将列表保存为CSV文件
with open('experiment_setup.csv', mode='w', newline='') as file:
    writer = csv.writer(file, delimiter=',')  
    for row in experiments:
        writer.writerow([''.join(map(str, row))])
```
